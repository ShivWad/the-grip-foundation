```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
dataset = pd.read_csv('SampleSuperstore.csv')
data = dataset
data.head()
```

```python
data.tail()
```

```python
data.dtypes
```

```python
data.nunique()
```

```python
data = data.drop(['Country'], axis =1 )
```

```python
features_with_na = [features for  features in data.columns if data[features].isnull().sum()>1 ]
features_with_na
```

```python
sns.boxplot(x=data['Profit'])
```

```python
sns.boxplot(x=data['Sales'])
```

```python
sns.regplot(data['Sales'], data['Profit'])
```

```python
sns.lineplot(data['Discount'], data['Sales'])
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Category", hue="Region", data=data)
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=data)
plt.show()
```

```python
Sub_category = np.array(set(data['Sub-Category']))
Sub_category
```
